```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
# print(iris.head()) # Display the first few rows of the dataset
# print(iris.describe()) # Display summary statistics of the dataset
# print(iris.info()) # Display information about the dataset
iris ['species'].value_counts() # Count occurrences of each species
 #print(iris['species'].value_counts()) # Display the counts of each species
```

```python
iris.describe()
```

```python
col = 'sepal_length' 
sns.histplot(data=iris, x="sepal_length", kde=True).set_title
(f'Distribuição da variável {col}') # Distribuição do valor menor ao maior
```

```python
for col in iris.drop(columns='species').columns:
    sns.histplot(data=iris, x=col, kde=True).set_title
    (f'Distribuição da variável {col}') 
    plt.show()
    
```

#interpretando os gráficos
# #Sepal Length: A maioria das flores tem um comprimento de sépala entre 5 e 7 cm, com um pico em torno de 6 cm.
# Sepal Width: A largura da sépala é mais uniformemente distribuída, com valores que variam de cerca de 2 a 4.5 cm.
# Petal Length: A maioria das flores tem um comprimento de pétala entre 1 e 6 cm, com um pico em torno de 5 cm. Há uma clara distinção entre as espécies, com Iris setosa tendo pétalas mais curtas.
# Petal Width: A largura da pétala também mostra uma distinção clara entre as espécies, com Iris setosa tendo pétalas mais estreitas.   A maioria das flores tem uma largura de pétala entre 0.1 e 2.5 cm.              

```python
if 'iris' not in locals():
    iris = sns.load_dataset("iris")

for col in iris.drop(columns='species').columns:
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(
        f'Distribuição da variável {col}'
    )
    plt.show()
    
```

```python
for col in iris.drop(columns='species'):
    sns.barplot(data=iris, x='species', y=col, hue='species')
    plt.title(f'Comparação da variável {col} por espécie')
    plt.xticks(rotation=90)
    plt.show()
    
```
