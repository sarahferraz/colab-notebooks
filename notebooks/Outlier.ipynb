```python
# Identificando outliers usando o método do desvio interquartil (IQR)
Q1 = houses['bathroom'].quantile(0.25) # Primeiro quartil (25º percentil)
Q3 = houses['bathroom'].quantile(0.75) # Terceiro quartil (75º percentil)
IQR = Q3 - Q1 # Intervalo interquartil

print(f'IQR {IQR}') # Imprime o valor do IQR
```

```python
houses_outliers = houses[(houses['bathroom'] < (Q1 - 1.5 * IQR)) | (houses['bathroom'] > (Q3 + 1.5 * IQR))]
# Filtra os outliers na coluna 'bathroom'
print(f'Outliers:\n{houses_outliers}') # Imprime os outliers encontrados
```
